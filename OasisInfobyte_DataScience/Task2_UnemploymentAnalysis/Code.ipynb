```python
#IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import plotly.express as px
```

```python
#LOAD DATASET
data_set = pd.read_csv("C:\\Users\\Mridul_Work\\Desktop\\OasisInfobyte_DataScience\\Task2_UnemploymentAnalysis\\Unemployment_Rate_upto_11_2020.csv")

#PRINT DATASET
print("UNEMPLOYMENT IN INDIA".center(0))
print(data_set)
print("\nDATASET INFO\n".center(0))
data_set.info()
print("\nDATASET DESCRIBED".center(0))
data_set.describe()
```

```python
#DATASET CLEANING
print("NULL VALUES: ",data_set.isnull().sum())
print("\nDUPLICATES :", data_set.duplicated().sum())
```

```python
#VISUALIZATION OF DATASET
#CORRELATION BETWEEN DATASET
sns.heatmap(data_set.corr(),cmap = "Blues")
plt.show()
```

```python
#PAIRPLOT OF COMPLETE DATA TO UNDERSTAND CORRELEATION
sns.pairplot(data_set)
```

```python
#HISTOGRAM REPRESENTATION OF DATASET
#changing name of columns for easy representation
data_set.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]

unemploment = data_set[["Estimated Unemployment Rate", "States", "Region"]]
figure = px.sunburst(unemploment, path=["Region", "States"], values="Estimated Unemployment Rate",title="INDIA'S UNEMPLOYMENT RATE - DIVIDING INTO ZONES",height=800, width=800)
print("\nPIE CHART")
figure.show()

plt.title("UNEMPLOYMENT IN INDIA")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data_set)
print("\nHISTOGRAM")
plt.show()
```
